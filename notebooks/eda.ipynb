```python
import pandas as pd
import numpy as np
```

```python
ufcstats_df = pd.read_json(r"C:\Development\ultimateNakMuay\data\raw\ufcstats.json")
```

```python
ufcstats_df["event_date"] = ufcstats_df["event_date"].apply(lambda x: pd.to_datetime(x))
```

```python
ufcstats_df[["event_title", "fighter1", "fighter2"]] = ufcstats_df['event_name'].str.split(':| vs\. ', expand=True)
```

```python
ufcstats_df = ufcstats_df.drop(columns="event_name")
```

```python
ufcstats_df[["city", "state", "country"]] = ufcstats_df['event_locations'].str.split(", ", expand = True)
```

```python
ufcstats_df.drop(columns="event_locations", axis=1, inplace= True)
```
